```python
# 2.2.2完善原始折线图1(辅助显示层)
# 需求:画出某城市11点到12点1小时内每分钟的温度变化折线图,温度范围在15度-18度
import random
import matplotlib.pyplot as plt
# 1. 准备数据
x = range(60)
y_shanghai = [random.uniform(15,18) for i  in x]

# 2. 创建画布
plt.figure(figsize=(16,8),dpi=80) # 

# 3. 绘制图像
plt.plot(x,y_shanghai)
# 3.1修改x,y刻度
# 3.11准备x刻度说明
x_label = ["11h{}m".format(i) for i in x]
plt.xticks(x[::5],x_label[::5])
# plt.yticks(range(0,40)[::5])
plt.yticks(range(0,40,5))

# 3.2 添加网格显示
plt.grid(linestyle="--",alpha=0.5)

# 3.3 添加描述信息
plt.xlabel("time")
plt.ylabel("temperature")
plt.title("City's temperature change 11h~12h")

# 4. 显示图像
plt.show()

```

```python
# 2.2.3完善原始折线图2(图像层)
# 需求:再添加一个城市的温度变化收集到beijing当天温度变化情况,温度在1度到3度.

import random
import matplotlib.pyplot as plt
# 1. 准备数据
x = range(60)
y_shanghai = [random.uniform(15,18) for i  in x]
y_beijing = [random.uniform(1,3) for i  in x]

# 2. 创建画布
plt.figure(figsize=(16,8),dpi=80) # 

# 3. 绘制图像
plt.plot(x,y_shanghai,color="r",linestyle="--",label="Shanghai") # 红色虚线
plt.plot(x,y_beijing,color="b",label="beijing")  # 添加第二组数据  蓝色实线

# 显示图例
plt.legend() 


# 3.1修改x,y刻度
# 3.11准备x刻度说明
x_label = ["11h{}m".format(i) for i in x]
plt.xticks(x[::5],x_label[::5])
# plt.yticks(range(0,40)[::5])
plt.yticks(range(0,40,5))

# 3.2 添加网格显示
plt.grid(linestyle="--",alpha=0.5)

# 3.3 添加描述信息
plt.xlabel("time")
plt.ylabel("temperature")
plt.title("Two City's temperature change 11h~12h")

# 4. 显示图像
plt.show()
```

```python
# 2.2.4多个坐标系显示-plt.subplots(面向对象的画图方法)
# 如果我们想要将上海和北京的天气图显示在同一个图的不同坐标系当中:

import random
import matplotlib.pyplot as plt
# 1. 准备数据
x = range(60)
y_shanghai = [random.uniform(15,18) for i  in x]
y_beijing = [random.uniform(1,3) for i  in x]

# 2. 创建画布
# plt.figure(figsize=(16,8),dpi=80) # 
figure, axes = plt.subplots(nrows=1, ncols=2, figsize=(32,8) ,dpi=80)

# 3. 绘制图像
# plt.plot(x,y_shanghai,color="r",linestyle="--",label="Shanghai") # 红色虚线
# plt.plot(x,y_beijing,color="b",label="beijing")  # 添加第二组数据  蓝色实线

axes[0].plot(x,y_shanghai,color="r",linestyle="--",label="Shanghai") # 红色虚线
axes[1].plot(x,y_beijing,color="b",label="beijing")  # 添加第二组数据  蓝色实线

# 显示图例
# plt.legend() 
axes[0].legend() 
axes[1].legend() 


# 3.1修改x,y刻度
# 3.11准备x刻度说明
x_label = ["11h{}m".format(i) for i in x]
# plt.xticks(x[::5],x_label[::5])
# plt.yticks(range(0,40,5))

axes[0].set_xticks(x[::5],x_label[::5])
axes[0].set_yticks(range(0,40,5))
axes[1].set_xticks(x[::5],x_label[::5])
axes[1].set_yticks(range(0,40,5))

# 3.2 添加网格显示
# plt.grid(linestyle="--",alpha=0.5)
axes[0].grid(linestyle="--",alpha=0.5)
axes[1].grid(linestyle="--",alpha=0.5)

# 3.3 添加描述信息
axes[0].set_xlabel("time")
axes[0].set_ylabel("temperature")
axes[0].set_title("shanghai City's temperature change 11h~12h")
axes[1].set_xlabel("time")
axes[1].set_ylabel("temperature")
axes[1].set_title("Beijing City's temperature change 11h~12h")

# 4. 显示图像
plt.show()

```

#  绘制柱状图(bar) 用于对比
•比较不同类别的数量或频率。
•展示分类数据的分布情况。

```python
import numpy as np
import random
import matplotlib.pyplot as plt

# 需求1-对比每部电影的票房收入

# 1. 准备数据
movie_titles = ["Galactic Quest",    "The Lost Kingdom",    "Future Storm",    "Endless Darkness",    "Mystic Realm",    "Hero's Return"]
box_office = [ 45.2,      38.7,      52.3,      29.8,      41.5,      34.2]

# 2. 创建画布
plt.figure(figsize=(16,8),dpi=80)

# 3. 绘制图像

x_ticks = range(len(movie_titles))
plt.bar(x_ticks, box_office, color=['r','b','g','y','c','k'])
# 3.1 添加网格显示
plt.grid(linestyle="--", alpha=0.5)
# 3.2 修改x刻度
plt.xticks(x_ticks,movie_titles)

plt.title("movie money ")
# 4. 显示图像
plt.show()
```

```python
# 需求2-如何对比电影票房收入才更能加有说服力?
# 比较相同天数的票房
import random
import matplotlib.pyplot as plt
# 1. 准备数据
movie_titles = ["Galactic Quest",    "The Lost Kingdom",    "Future Storm"]
first_day = [1125,2512,1252]
first_week = [3596,8541,4521]


# 2. 创建画布
plt.figure(figsize=(16,8),dpi=80)

# 3. 绘制图像
x_ticks = range(len(movie_titles))
plt.bar(x_ticks, first_day, color=['r','b','g','y','c','k'], width=0.2, label="first_day")
plt.bar([x+0.2 for x in x_ticks], first_week, color=['b','r','y','c','b','r'], width=0.2, label="first_week")

# 3.1 添加网格显示
plt.grid(linestyle="--", alpha=0.5)
# 3.2 修改x刻度 
plt.xticks([x+0.1 for x in x_ticks], movie_titles)

plt.legend() # show 图例

plt.title("movie money ")
# 4. 显示图像
plt.show()

```

#  绘制直方图(histogram)
•展示连续数据的分布情况。
•观察数据的集中趋势和分布形态。

·组数:在统计数据时,我们把数据按照不同的范围分成几个组,分成的组的个数称为组数
·组距:每一组两个端点的差

1.直方图展示数据的分布,柱状图比较数据的大小.
2.直方图X轴为定量数据,柱状图X轴为分类数据.
3.直方图柱子无间隔,柱状图柱子有间隔
4.直方图柱子宽度可不一,柱状图柱子宽度须一致


```python
import random
import matplotlib.pyplot as plt
#需求:电影时长分布状况


# 1. 准备数据
time = [131,98,125,131,124,139,121,117,128,109,136,135,124,102,100,104,114,115,114,158,105]
# bins 组数 = (max(time)-min(time)) // 组距

# 2. 创建画布
plt.figure(figsize=(16,8), dpi=80)

distance = 2 
bins = int((max(time)-min(time)) / distance)

# 3. 绘制图像
plt.hist(time,bins=bins,edgecolor='k')

# 3.1 添加网格显示
plt.grid(linestyle="--", alpha=0.5)
# 3.2 修改x刻度 
plt.xticks(range(min(time),max(time)+distance,distance))
# 3.2 show 图例

# 4. 显示图像
plt.show()
```

#  绘制饼图(pie)
•展示各部分占总体的比例。
•强调各个部分相对于整体的重要性。




```python
import random
import matplotlib.pyplot as plt
#需求:显示不同的电影的排片占比


# 1. 准备数据
movie = ["11","22","33","44","55"]
place_count = [6255,2542,3652,6365,1251]
# 2. 创建画布
plt.figure(figsize=(16,8), dpi=80)


# 3. 绘制图像
plt.pie(place_count,labels=movie,autopct="%1.2f  %%")

plt.axis('equal')
# 3.2 show 图例
plt.legend()

# 4. 显示图像
plt.show()
```
